**2. Load datasets**

```python
# Read the data
Books = pd.read_csv('dataset/bx_271380_books.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
Users = pd.read_csv('dataset/bx_278860_users.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
Ratings = pd.read_csv('dataset/bx_1149781_book_ratings.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
```

```python
name_ratings = pd.merge(Books, Ratings, on='ISBN')
```

```python
name_ratings.head(4)
```

```python
x = name_ratings.groupby('User-ID').count()['Book-Rating'] > 200
user_rating_count_df = x[x].index
```

```python
user_rating_count_df
```

```python
filtered_ratings = name_ratings[name_ratings['User-ID'].isin(user_rating_count_df)] 
```

```python
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
popular_books_df = y[y].index
```

```python
popular_books_df
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books_df)]
```

```python
final_ratings
```

**Let's use the pivot table to display the (User-ID, Book-Title(Book-Rating))**

```python
user_book_rating_pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
```

```python
user_book_rating_pt
```

## Implementation of similarity using scikit-learn library

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(user_book_rating_pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_title, recommendation_count):
    #if the book is not in the dataset, return
    if book_title not in user_book_rating_pt.index:
        print("Sorry!!! this book does not exist in the dataset!")
        return
    book_index = user_book_rating_pt.index.get_loc(book_title)
    similar_books = list(enumerate(similarity_scores[book_index]))
    similar_books = sorted(similar_books, key=lambda x:x[1], reverse=True)
    similar_books = similar_books[1:recommendation_count+1]
    # book_indices = [i[0] for i in similar_books]
    #return user_book_rating_pt.index[book_indices]
    # Print the names of the books with ordered list
    print(f'The top {recommendation_count} books similar to "{book_title}" are:' )
    i = 1
    for book in similar_books:
        print(str(i) + '. ' + user_book_rating_pt.index[book[0]])
        i += 1
```

```python
recommend('The Lovely Bones: A Novel', 10)
```

```python
book_name = input("Enter the name of the book: ")
recommendation_count = int(input("Enter the number of recommendations you want: "))
recommend(book_name, recommendation_count)
```
