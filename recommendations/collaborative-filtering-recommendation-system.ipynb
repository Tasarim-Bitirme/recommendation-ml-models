**2. Load datasets**

```python
# Read the data
Books = pd.read_csv('dataset/bx_271380_books.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
Users = pd.read_csv('dataset/bx_278860_users.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
Ratings = pd.read_csv('dataset/bx_1149781_book_ratings.csv', on_bad_lines='skip', sep=';', encoding='latin-1')
```

```python
Ratings.head()
```

```python
name_ratings = pd.merge(Books, Ratings, on='ISBN')
```

```python
name_ratings.head(4)
```

```python
x = name_ratings.groupby('User-ID').count()['Book-Rating'] > 200
user_rating_count_df = x[x].index
```

```python
user_rating_count_df
```

```python
filtered_ratings = name_ratings[name_ratings['User-ID'].isin(user_rating_count_df)] 
```

```python
filtered_ratings
```

```python
#Take the first 50 books with the highest rating count
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
#Create a new dataframe with the books that have a rating count of 50 or more
popular_books_df = y[y].index
```

```python
popular_books_df
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books_df)]
```

```python
final_ratings
```

**Let's use the pivot table to display the (User-ID, ISBN, Book-Title(Book-Rating))**

```python
#Create a pivot table with the user id as the index, the book isbn and title as the columns and the book rating as the values
user_book_rating_pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
```

```python
#form a table with book and user rating values as columns and rows respectively
user_book_rating_pt
```

```python
book_isbn_title_user_rating_pt = final_ratings.pivot_table(index=['ISBN','Book-Title'], columns='User-ID', values='Book-Rating').fillna(0)
```

```python
book_isbn_title_user_rating_pt.head(4)
```

```python
#Print the book isbn only:
book_isbn_title_user_rating_pt.index.get_level_values(0)
```

## Implementation of similarity using scikit-learn library

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(book_isbn_title_user_rating_pt)
```

```python
similarity_scores.shape
```

```python
similarity_scores[0]
```

```python
#Sort the similarity scores in descending order
sorted(list(enumerate(similarity_scores[0])), reverse=True, key=lambda x:x[1])[1:11]  
```
